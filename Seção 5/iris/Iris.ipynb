```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense # camada densa
from tensorflow.keras import utils as np_utils #funçoes utilitarias
from sklearn.preprocessing import LabelEncoder # biblioteca padrao de ML
from sklearn.model_selection import train_test_split #para fazer a divisao da base de dados em teste e treinamento
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv(r'C:\Users\thier\Jupyter\Card-12\Seção 5\Iris.csv')
```

```python
base
```

Seleciona todas as linhas do banco de dados e as colunas 0, 1, 2, 3

```python
X = base.iloc[:, 0:4].values
```

```python
X
```

```python
y = base.iloc[:, 4].values
```

```python
y
```

Transformar a string em numeros: Iris-setosa=0, Iris-versicolor=1, Iris-virginic=2

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
y
```

Transformando esses valores em matriz

```python
y = np_utils.to_categorical(y)
y
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.25)
```

```python
X_treinamento.shape, X_teste.shape
```

```python
y_treinamento.shape, y_teste.shape
```

# Arquitetura da Rede Neural

```python
#Criar uma sequencia de camadas
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),#camada aculta
    tf.keras.layers.Dense(units = 4, activation = 'relu'),#camada aculta
    tf.keras.layers.Dense(units = 3, activation = 'softmax'),#camada final
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy']) #compilar a rede
```

```python
rede_neural.fit(X_treinamento, y_treinamento, batch_size = 10, epochs = 1000)
```

# Avaliação e Previsões 

```python
rede_neural.evaluate(X_teste, y_teste)
```

Probabilidade para cada um dos neurônios

```python
previsoes = rede_neural.predict(X_teste)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
y_teste2 = [np.argmax(t) for t in y_teste]
print(y_teste2)
```

```python
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
```

Taxa de acerto:

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste2, previsoes2)
```

```python
confusion_matrix(y_teste2, previsoes2)
```
